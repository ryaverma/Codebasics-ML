```python
%matplotlib
import matplotlib.pyplot as plt
```

```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
dir(iris)
```

### Creating a DataFrame

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

### Adding Additional Columns

```python
df["target"] = iris.target
df.head()
```

```python
df["flower_name"] = df.target.apply(lambda x: iris.target_names[x])
df
```

The .apply() function is used to apply a function to each element of the df.target column. <br>
lambda x: iris.target_names[x] is an anonymous function that takes each value x (an integer label) from the target column and maps it to the corresponding flower name using the iris.target_names array.

### Data Visualization

```python
df0 = df[df.target==0]
df1 = df[df.target==1]
df2 = df[df.target==2]
```

```python
plt.xlabel("sepal length (cm)")
plt.ylabel("sepal width (cm)")
plt.scatter(df0["sepal length (cm)"], df0["sepal width (cm)"], color="green", marker="+")
plt.scatter(df1["sepal length (cm)"], df1["sepal width (cm)"], color="blue", marker=".")
```

```python
plt.xlabel("petal length (cm)")
plt.ylabel("petal width (cm)")
plt.scatter(df0["petal length (cm)"], df0["petal width (cm)"], color="green", marker="+")
plt.scatter(df1["petal length (cm)"], df1["petal width (cm)"], color="blue", marker=".")
```

### SVM

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop(["target", "flower_name"], axis="columns")
```

```python
y = df.target
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
from sklearn.svm import SVC
model = SVC()
```

```python
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```
