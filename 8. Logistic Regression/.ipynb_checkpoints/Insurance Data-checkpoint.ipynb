```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("insurance_data.csv")
df
```

```python
plt.scatter(df.age, df.bought_insurance, marker="+", color="red")
```

### Train and Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(df[["age"]], df.bought_insurance, train_size=0.9)
```

```python
x_test
```

```python
x_train
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
reg = LogisticRegression()
```

```python
reg.fit(x_train, y_train)
```

```python
reg.predict(x_test)
```

```python
reg.score(x_test, y_test)
```

### Impact of Salary on Employee Retention

```python
pd.crosstab(hrdf.salary, hrdf.left).plot(kind="bar")
```

In pandas, .crosstab() is a function used to compute a cross-tabulation (or contingency table) of two or more factors. It is often used to summarize the relationship between categorical variables, showing the frequency of combinations of different categories.

pandas.crosstab(index, columns, values=None, aggfunc=None, **kwargs)

Parameters:
index: The row variable(s) you want to group by. <br>
columns: The column variable(s) you want to group by. <br>
values (optional): An array of values you want to aggregate, often used with aggfunc. <br>
aggfunc (optional): The aggregation function to apply to the values (e.g., sum, mean). <br>
**kwargs: Other optional arguments such as margins for adding row/column totals.

### Department Wise Employee Retention Rate

```python
pd.crosstab(hrdf.Department, hrdf.left).plot(kind="bar")
```


From above chart there seem to be some impact of department on employee retention but it is not major hence we will ignore department in our analysis

### From the Data Analysis so far we can conclude that we will use following variables as independant variables in our model

**Satisfaction Level** <br>
**Average Monthly Hours** <br>
**Promotion Last 5 Years** <br>
**Salary**

```python
subhrdf = hrdf[["satisfaction_level", "average_montly_hours", "promotion_last_5years", "salary"]]
subhrdf
```

### Dummy Variable Method for Salary Column

```python
dummies = pd.get_dummies(subhrdf.salary, dtype=int)
merged = pd.concat([subhrdf, dummies], axis="columns")
merged
```

```python
merged.drop("salary", axis="columns", inplace=True)
merged
```

### Logistic Regression

```python
x = merged
y = hrdf.left
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3)
```

```python
reg = LogisticRegression()
reg.fit(x_train, y_train)
```

```python
reg.predict(x_test)
```

```python
reg.score(x_test, y_test)
```
