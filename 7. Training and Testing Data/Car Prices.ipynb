```python
%matplotlib
import matplotlib.pyplot as plt
```

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_csv("carprices.csv")
df
```

### Visualization to Understand the Relationship between Car Mileage and Selling Price

```python
plt.scatter(df["Mileage"], df["Sell Price($)"])
```

### Visualization to Understand the Relationship between Car Age and Selling Price

```python
plt.scatter(df["Age(yrs)"], df["Sell Price($)"])
```

```python
x = df.drop(["Sell Price($)"], axis="columns")
x
```

```python
y = df["Sell Price($)"]
y
```

### Training and Testing

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

If you want to get constant train and test data every time the above statement is executed use random_state=1 <br>
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=1)

### Training the Model

```python
reg = linear_model.LinearRegression()
reg.fit(x_train, y_train)
```

### Testing the Model

```python
reg.predict(x_test)
```

### Accuracy

```python
reg.score(x_test, y_test)
```
