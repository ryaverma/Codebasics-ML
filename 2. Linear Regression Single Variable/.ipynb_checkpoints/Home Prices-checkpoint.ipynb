```python
%matplotlib
import matplotlib.pyplot as plt
```

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df = pd.DataFrame({"area":[2600,3000,3200,3600,4000], "prices":[550000,565000,610000,680000,725000]})
```

```python
df
```

```python
plt.xlabel("Area (Sqr ft)")
plt.ylabel("Prices (USD)")
plt.scatter(df.area,df.prices,color="red",marker="+")
```

### Creating Linear Regression Object

```python
reg = linear_model.LinearRegression()
reg.fit(df[["area"]],df.prices)
```

The reason we use a 2D array for the input features (df[["area"]]) is because the fit() method in sklearn.linear_model.LinearRegression() expects the input data (the feature matrix) to be in the form of a 2D array, even if there's only one feature (as in this case with just area).

```python
reg.predict([[3300]])
```

```python
reg.predict([[5000]])
```

### Slope and Intercept to Find Dependent Variable

In linear equations
y = mx + b (here m=slope and b=y intercept)

Similarly here,
price = m*area + b

```python
reg.coef_
```

```python
reg.intercept_
```

### Saving Trained Model with Pickle

```python
import pickle
```

```python
with open("homeprices_pickle","wb") as f:
    pickle.dump(reg,f)
```

```python
with open("homeprices_pickle", "rb") as f:
    preg = pickle.load(f)
```

```python
preg.predict([[5000]])
```

### Saving Trained Model using Joblib

joblib is used instead of pickle when we have large numpy arrays

```python
import joblib
```

```python
joblib.dump(reg, "homeprices_joblib")
```

```python
jreg = joblib.load("homeprices_joblib")
```

```python
jreg.predict([[5000]])
```
