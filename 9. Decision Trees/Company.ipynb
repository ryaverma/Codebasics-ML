```python
import pandas as pd
df = pd.read_csv("salaries.csv")
df
```

inputs is Independent Variables and target is the Dependent Variable

```python
inputs = df.drop("salary_more_then_100k", axis="columns")
target = df["salary_more_then_100k"]
```

### Using Label Encoder to Convert Company, Job and Degree

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

### Adding New Columns

```python
inputs["company_n"] = le.fit_transform(inputs["company"])
inputs["job_n"] = le.fit_transform(inputs["job"])
inputs["degree_n"] = le.fit_transform(inputs["degree"])
```

```python
inputs
```

```python
inputs_n = inputs.drop(["company", "job", "degree"], axis="columns")
inputs_n
```

### Training the Classifier

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
model.fit(inputs_n, target)
```

```python
model.predict([[2,0,0]])
```
